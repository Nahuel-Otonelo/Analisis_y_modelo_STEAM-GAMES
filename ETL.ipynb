### 5- Columna 'realease_date'

El procedimiento de limpieza de esta columna es algo distinto a las otras. Primero cambio de tipo de dato, luego cambio algunos de los valores que no son del tipo año-mes-dia como por ejemplo 2020-10-23 por un valor de fecha 1900-01-01. Algunos otros como por ejemplo 'Jun 2009' los remplazo por '2009-06-01'.

```python
juegos['release_date']=juegos['release_date'].astype(str)
juegos['release_date'].replace(to_replace='0000-01-01', value= '1900-01-01', inplace=True)
juegos['release_date'].replace(to_replace='Soon..', value= '1900-01-01', inplace=True)
juegos['release_date'].replace(to_replace='Beta测试已开启', value= '1900-01-01', inplace=True)
juegos['release_date'].replace(to_replace='Jun 2009', value= '2009-06-01', inplace=True)
juegos['release_date'].replace(to_replace='Oct 2010', value= '2010-10-01', inplace=True)
juegos['release_date'].replace(to_replace='Feb 2011', value= '2011-02-01', inplace=True)
juegos['release_date'].replace(to_replace='Aug 2014', value= '2014-08-01', inplace=True)
juegos['release_date'].replace(to_replace='Sep 2014', value= '2014-09-01', inplace=True)
juegos['release_date'].replace(to_replace='Apr 2015', value= '2015-04-01', inplace=True)
juegos['release_date'].replace(to_replace='Apr 2016', value= '2016-04-01', inplace=True)
juegos['release_date'].replace(to_replace='Jul 2016', value= '2016-07-01', inplace=True)
juegos['release_date'].replace(to_replace='June 2016', value= '2016-05-01', inplace=True)
juegos['release_date'].replace(to_replace='Coming Soon', value= '1900-01-01', inplace=True)
juegos['release_date'].replace(to_replace='Q2 2017', value= '2017-04-01', inplace=True)
juegos['release_date'].replace(to_replace='TBA', value= '1900-01-01', inplace=True)
juegos['release_date'].replace(to_replace= "When it's", value= '1900-01-01', inplace=True)
juegos['release_date'].replace(to_replace= "coming soon", value= '1900-01-01', inplace=True)

###############################
# esto podria haberse hecho de alguna manera mas eficiente.... para otra oportunidad...
###############################
```

Luego de estos remplazos, cambio el tipo de dato mediante datetime. Utilizo el parametro errors= 'coerce' para que los valores que no pueden transformarse o que dan error, devuelvan el valor NaT.

```python
juegos['release_date']=pd.to_datetime(juegos['release_date'], errors= 'coerce', format= '%Y/%m/%d')
```

Ahora paso a tipo de dato str nuevamente y cambio nos 'NaT' por el mismo valor de fecha por defecto '1900-01-01' y luego vuelvo a cambiar al tipo datetime, quedando asi la columna limpia sin errores.


```python
juegos['release_date']=juegos['release_date'].astype('str')
juegos['release_date'].replace(to_replace= 'NaT', value= '1900-01-01', inplace=True)
```

```python
juegos['release_date']=pd.to_datetime(juegos['release_date'], errors= 'coerce', format= '%Y/%m/%d')
```
