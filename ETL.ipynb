```python
import pandas as pd
import numpy as np
import ast
```

```python
filas=[]
with open('steam_games.json') as a:
    for linea in a.readlines():
        filas.append(ast.literal_eval(linea))
juegos=pd.DataFrame(filas)
```

```python
juegos['app_name']=juegos['app_name'].astype('str')
juegos['app_name'].replace(to_replace= 'nan' , value='no_info', inplace=True)
```

#### Conclusion: 
Los registros 'no_info' de 'app name', son 'NaN' en 'title'.

Ahora buscaremos en que se diferencian title de app_name, cuando los valores de los registros son distintos.

```python
juegos['title']=juegos['title'].astype('str')
juegos['title'].replace(to_replace= 'nan' , value='no_info', inplace=True)
```

Como se ve en la tabla anterior, las diferencias vienen de que en la columna 'title' despues del & le sigue un 'amp;'. Parece ser un error de compilacion o de formatos. Para verificarlo, vemos cuantas registros de la columna 'title' contienen un '&':

```python

juegos.loc[juegos['title'].str.contains('&', regex=True)]
```

### Se elimina la columna 'title'
Puede verse que solo en 3 registros, la diferencia entre 'title' y 'app_name' no es por un NaN, ni por el error del '&amp'.
Por todo esto se decide que se elimina la columna title ya que como se demostro, no aporta informacion alguna.

```python
juegos=juegos.drop(['title'], axis=1)
```
